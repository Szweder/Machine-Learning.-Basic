```python
#–ó–∞–¥–∞—á–∞ 1 ‚Äî –•–ª–µ–± –¥–ª—è —Ö–æ–±–±–∏—Ç–æ–≤

def share_bread(N, K):
    """
    –î–µ–ª–∏—Ç K –∫—É—Å–∫–æ–≤ —Ö–ª–µ–±–∞ –Ω–∞ N —Ö–æ–±–±–∏—Ç–æ–≤.
    –í–æ–∑–≤—Ä–∞—â–∞–µ—Ç:
    - x: —Å–∫–æ–ª—å–∫–æ –∫—É—Å–∫–æ–≤ –¥–æ—Å—Ç–∞–Ω–µ—Ç—Å—è –∫–∞–∂–¥–æ–º—É —Ö–æ–±–±–∏—Ç—É
    - y: —Å–∫–æ–ª—å–∫–æ –æ—Å—Ç–∞–Ω–µ—Ç—Å—è –≤ –∫–æ—Ä–∑–∏–Ω–∫–µ —É –°—ç–º–∞
    """
    x = K // N
    y = K % N
    return x, y

# –ü—Ä–æ–≤–µ—Ä–∫–∞
assert share_bread(N=3, K=14) == (4, 2)
```

```python
#–ó–∞–¥–∞—á–∞ 2 ‚Äî –í–∏—Å–æ–∫–æ—Å–Ω—ã–π –≥–æ–¥

def leap_year(year):
    """
    –ü—Ä–æ–≤–µ—Ä—è–µ—Ç, —è–≤–ª—è–µ—Ç—Å—è –ª–∏ –≥–æ–¥ –≤–∏—Å–æ–∫–æ—Å–Ω—ã–º.
    –í–æ–∑–≤—Ä–∞—â–∞–µ—Ç 'YOU SHALL PASS' –µ—Å–ª–∏ –¥–∞, –∏–Ω–∞—á–µ 'YOU SHALL NOT PASS'
    """
    if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
        return 'YOU SHALL PASS'
    else:
        return 'YOU SHALL NOT PASS'

# –ü—Ä–æ–≤–µ—Ä–∫–∞
assert leap_year(5) == 'YOU SHALL NOT PASS'
```

```python
#–ó–∞–¥–∞—á–∞ 3 ‚Äî –ü–ª–æ—â–∞–¥—å –∞–º—É–ª–µ—Ç–∞ –ì–µ–Ω–¥–∞–ª—å—Ñ–∞

def amulet_area(a, b, c):
    """
    –°—á–∏—Ç–∞–µ—Ç –ø–ª–æ—â–∞–¥—å —Ç—Ä–µ—É–≥–æ–ª—å–Ω–∏–∫–∞ –ø–æ —Ñ–æ—Ä–º—É–ª–µ –ì–µ—Ä–æ–Ω–∞.
    """
    s = (a + b + c) / 2
    S = (s * (s - a) * (s - b) * (s - c)) ** 0.5
    return S

# –ü—Ä–æ–≤–µ—Ä–∫–∞
assert amulet_area(3, 4, 5) == 6
```

```python
#–ó–∞–¥–∞—á–∞ 4 ‚Äî –ú–µ—Ç—Ä–∏–∫–∏ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–π –¥–æ –ú–æ—Ä–¥–æ—Ä–∞

import numpy as np

def cal_euclidean(a, b):
    """
    –ï–≤–∫–ª–∏–¥–æ–≤–æ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ –º–µ–∂–¥—É –¥–≤—É–º—è –≤–µ–∫—Ç–æ—Ä–∞–º–∏
    """
    return sum((x - y) ** 2 for x, y in zip(a, b)) ** 0.5

def cal_manhattan(a, b):
    """
    –ú–∞–Ω—Ö—ç—Ç—Ç–µ–Ω—Å–∫–æ–µ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ –º–µ–∂–¥—É –¥–≤—É–º—è –≤–µ–∫—Ç–æ—Ä–∞–º–∏
    """
    return sum(abs(x - y) for x, y in zip(a, b))

def cal_cosine(a, b):
    """
    –ö–æ—Å–∏–Ω—É—Å–Ω–æ–µ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ –º–µ–∂–¥—É –¥–≤—É–º—è –≤–µ–∫—Ç–æ—Ä–∞–º–∏
    """
    dot = sum(x * y for x, y in zip(a, b))
    norm_a = sum(x ** 2 for x in a) ** 0.5
    norm_b = sum(y ** 2 for y in b) ** 0.5
    return 1 - dot / (norm_a * norm_b)
```

```python
# –ü—Ä–∏–º–µ—Ä —Ä–∞–±–æ—Ç—ã —Å –º–µ—Ç—Ä–∏–∫–∞–º–∏
a = np.random.randint(-10, 10, size=10)
b = np.random.randint(-10, 10, size=10)

print("a =", a)
print("b =", b)
print("üìê Euclidean:", cal_euclidean(a, b))
print("üèôÔ∏è Manhattan:", cal_manhattan(a, b))
print("üß≠ Cosine:", cal_cosine(a, b))
```
